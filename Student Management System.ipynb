```python
#class student
class student:
    def __init__(self, mssv, name, birth, class_stu,score,rate):
        self.mssv = mssv
        self.name = name
        self.birth = birth
        self.class_stu = class_stu
        self.score = score
        self.rate = rate
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import *
from tkinter import ttk

manage_system = Tk ()
manage_system .title ('Student management system')
manage_system.geometry("1150x600")

#List
list_stu = []

#TreeView
col =('mssv','name','birth','class_stu','score','rate')
tr = Treeview(manage_system,columns=col,show='headings')
tr.heading('mssv',text='MSSV')
tr.heading('name',text='Name')
tr.heading('birth',text='Birth')
tr.heading('class_stu',text='Class')
tr.heading('score',text='Score')
tr.heading('rate',text='Rate')
#Create Scrollbar
scrollx = Scrollbar(manage_system,orient=HORIZONTAL)
scrolly = Scrollbar(manage_system,orient=VERTICAL)
# Create label
mssv_lbl = Label(manage_system, text = 'MSSV:')
name_lbl = Label(manage_system, text = 'Name:')
birth_lbl = Label(manage_system, text = 'Birth:')
class_lbl = Label(manage_system, text = 'Class:')
score_lbl = Label(manage_system, text = 'Score:')

# Create entry
mssv = Entry(manage_system)
name = Entry(manage_system)
birth = Entry(manage_system)
class_stu = Entry(manage_system)
score = Entry(manage_system)
check = Entry(manage_system)

# Funtion
def rating():
    temp = int(mssv.get())
    selected = tr.focus()
    for stu in list_stu:
        if temp == stu.mssv:
            rate_score = float(stu.score)
            if rate_score >=8:
                rate_stu = 'Giỏi'
            elif rate_score < 8 and rate_score >= 6.5:
                rate_stu = 'Khá'
            elif rate_score < 6.5 and rate_score >= 5:
                rate_stu = 'Trung Bình'
            else:
                rate_stu = 'Yếu'
            stu.rate = rate_stu
    tr.item(selected, values=(mssv.get(), name.get(),birth.get(),class_stu.get(),score.get(),rate_stu))
    # Clear entry boxes
    mssv.delete(0, END)
    name.delete(0, END)
    birth.delete(0, END)
    class_stu.delete(0, END)
    score.delete(0, END)
    
def rating_2(rate_score):
    # Use rating_2 in funtion "update_stu"
    rate_score = float(rate_score)
    if rate_score >=8:
        rate_stu = 'Giỏi'
    elif rate_score < 8 and rate_score >= 6.5:
        rate_stu = 'Khá'
    elif rate_score < 6.5 and rate_score >= 5:
        rate_stu = 'Trung Bình'
    else:
        rate_stu = 'Yếu'
    return rate_stu

def add_stu():
    list_stu.append(student(int(mssv.get()), name.get(),birth.get(),class_stu.get(),float(score.get()),'?'))
    tr.insert('',END,values=(mssv.get(), name.get(),birth.get(),class_stu.get(),score.get(),'?'))
    print([stu.mssv for stu in list_stu])
    messagebox.showinfo("Student", mssv.get() + " added sucessfully")

def select_stu():
    # Clear entry boxes
    mssv.delete(0, END)
    name.delete(0, END)
    birth.delete(0, END)
    class_stu.delete(0, END)
    score.delete(0, END)
    # Grab record number
    selected = tr.focus()
    # Grab record values
    values = tr.item(selected, 'values')
    # output to entry boxes
    mssv.insert(0, values[0])
    name.insert(0, values[1])
    birth.insert(0, values[2])
    class_stu.insert(0, values[3])
    score.insert(0, values[4])

def update_stu():
    temp = int(mssv.get())
    # Grab record number
    selected = tr.focus()
    values = tr.item(selected, 'values')
    # Save new data
    tr.item(selected, values=(mssv.get(), name.get(),birth.get(),class_stu.get(),score.get(),rating_2(score.get())))
    for stu in list_stu:
        if temp == stu.mssv:
            stu.name = name.get()
            stu.birth = birth.get()
            stu.class_stu = class_stu.get()
            stu.score = score.get()
    # Clear entry boxes
    mssv.delete(0, END)
    name.delete(0, END)
    birth.delete(0, END)
    class_stu.delete(0, END)
    score.delete(0, END)
    
def clicker(e):
    select_stu()
    
def search_stu():
    search_mssv = int(check.get())
    exist = False 
    for stu in list_stu:
        if stu.mssv == search_mssv:
            exist = True
            messagebox.showinfo("Student", f"Information of {stu.mssv} is: \nName: {stu.name} \nBirth: {stu.birth} \nClass: {stu.class_stu} \nScore: {stu.score} \nRate: {stu.rate}")
            break
    if exist == False:
        messagebox.showinfo("Student", f"{search_mssv} does not exist in list")
        
def delete_stu():
    delete_mssv = int(mssv.get())
    x = tr.selection()[0]
    tr.delete(x)
    exist = False
    for i in range(0,len(list_stu)):
        if list_stu[i].mssv == delete_mssv:
            exist = True
            del list_stu[i]
            messagebox.showinfo("Student", f"{delete_mssv} is deleted successfully")
            break
    if exist == False:
        messagebox.showinfo("Student", f"{delete_mssv} does not exist in list student")
def export():
    file = open('student.txt','w')
    for stu in list_stu:
        entry = f"MSSV: {stu.mssv} \nName: {stu.name} \nBirth: {stu.birth} \nClass: {stu.class_stu} \nScore: {stu.score} \nRate: {stu.rate};\n"
        file.write(entry)
    file.close()
#Button
add_btn = Button(manage_system, text = 'Add', command=add_stu)
delete_btn = Button(manage_system, text = 'Delete', command=delete_stu)
reset_btn = Button(manage_system, text = 'Reset')
search_btn = Button(manage_system, text = 'Search MSSV', command=search_stu)
rate_btn = Button(manage_system, text = 'Rate', command=rating)
select_btn = Button(manage_system, text="Select", command=select_stu)
update_btn = Button(manage_system, text = 'Update', command=update_stu)
export_btn = Button(manage_system, text = 'Export to txt', command=export)
#Place
mssv_lbl.place(x=40,y=20)
name_lbl.place(x=40,y=50)
birth_lbl.place(x=40,y=80)
class_lbl.place(x=40,y=110)
score_lbl.place(x=40,y=140)

mssv.place(x=100,y=20)
name.place(x=100,y=50)
birth.place(x=100,y=80)
class_stu.place(x=100,y=110)
score.place(x=100,y=140)

add_btn.place(x=10,y=170)
delete_btn.place(x=100,y=170)
search_btn.place(x=30,y=200)
rate_btn.place(x=190,y=170)
#select_btn.place(x=?,y=?)
update_btn.place(x=280,y=170)
check.place(x=120,y=200)
export_btn.place(x=680,y=300)
tr.place(x=380,y=30,width=700)

#Scrollbar
scrollx.place(x=380,y=260,width=550)
tr.configure(xscrollcommand=scrollx.set)
tr.configure(selectmode='extended')
scrollx.configure(command=tr.xview)

#Bind
tr.bind("<ButtonRelease-1>", clicker)


manage_system.mainloop()
```
